## Metodologia CRISP-DM
CRISP-DM, que son las siglas de Cross-Industry Standard Process for Data Mining, es un método probado para orientar sus trabajos de minería de datos.

Como metodología, incluye descripciones de las fases normales de un proyecto, las tareas necesarias en cada fase y una explicación de las relaciones entre las tareas.
Como modelo de proceso, CRISP-DM ofrece un resumen del ciclo vital de minería de datos.

https://www.ibm.com/docs/es/spss-modeler/saas?topic=dm-crisp-help-overview

Lo primero que debe hacer cuando obtiene un nuevo conjunto de datos es echar un vistazo a algunos de ellos. Esto permite ver que todo se leyó correctamente y le da una idea de lo que está pasando con los datos. En este caso, veamos si faltan valores, que se representarán con NaN o Ninguno.

Observe las primeras cinco filas del archivo, se puede ver un puñado de datos faltantes! 🔼🔼

¿Parece que faltan algunos valores?

Bien, ahora se sabe que faltan algunos valores. Veamos cuántos tenemos en cada columna

```python
# obtener el número de puntos de datos faltantes por columna
missing_values_count = df.isnull().sum()
# mire el número de puntos que faltan en las primeras diez columnas
missing_values_count
```

¡Eso parece mucho! Podría ser útil ver qué porcentaje de los valores en nuestro conjunto de datos faltaba para darnos una mejor idea de la escala de este problema:

```python
# how many total missing values do we have?
total_cells = np.product(df.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

¡Menos de una cuarta parte de las celdas en este conjunto de datos están vacías! En el siguiente paso, se echara un vistazo más de cerca a algunas de las columnas a las que les faltan valores Y se intentara averiguar qué podría estar pasando con ellas.

```python
# eliminar todas las filas que contienen un valor faltantes
df.dropna()
```

Esto ha eliminado todos los datos! 😱 Esto se debe a que a cada fila de nuestro conjunto de datos le faltaba al menos un valor. Podríamos tener más suerte eliminando todas las columnas que tienen al menos un valor faltante.
